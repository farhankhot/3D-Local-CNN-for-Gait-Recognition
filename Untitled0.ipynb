```python
def train_model(model, criterion, optimizer, num_epochs):
  losses = []
  running_loss = 0

  # for i, inp in enumerate(dataloaders["train"]):
  #     inputs, labels = inp
  #     print("labels", labels)

  for epoch in range(num_epochs):
    # for i, inp in enumerate(dataloaders["train"]):
    for i, inp in enumerate(trainloader):

      # print("inp:", inp)
      # print("type of inp:", type(inp))


      inputs, labels = inp
      # print("label type:", type(labels))
      inputs, labels = inputs.to("cuda"), labels.to("cuda")
      optimizer.zero_grad()

      outputs = model(inputs)
      labels = labels.squeeze()
      print(labels)
      loss = criterion(outputs, labels)

      losses.append(loss.item())

      loss.backward()
      optimizer.step()

      running_loss += loss.item()

      if i%100 == 0 and i > 0:
        print(f'Loss [{epoch+1}, {i}](epoch, minibatch): ', running_loss / 100)
        running_loss = 0.0

    avg_loss = sum(losses)/len(losses)
    scheduler.step(avg_loss)
            
  print('Training Done')
```

```python
# ==================== GETTING FROM GOOGLE COLAB ========================= 
# train_dir = "gdrive/My Drive/Colab Notebooks/jersey numbers dataset 224"
# ==================== GETTING FROM GOOGLE COLAB =========================

train_dir = "jersey numbers dataset 224"
test_dir = "test jersey numbers dataset w,h more than 100"

transform_train = transforms.Compose([
  # transforms.RandomHorizontalFlip(),
  transforms.Resize( (100, 100) ),
  # transforms.RandomCrop(100, padding=4),
  transforms.ToTensor(),
  transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
])

def generate_image_paths(dir):

  # train_image_paths = []
  # for folders in os.listdir(train_dir):
  #   for files in os.listdir(train_dir + '/' + folders):
  #     #print(files)
  #     train_image_paths.append(train_dir+'/'+folders+'/'+files)

  # print(train_image_paths[3])

  image_paths = []
  for folders in os.listdir(dir):
    for files in os.listdir(dir + '/' + folders):
      #print(files)
      image_paths.append(dir+'/'+folders+'/'+files)

  return image_paths, len(image_paths)

def generate_classes_from_directory(dir):
  # classes=[]
  # for folders in os.listdir(train_dir):
  #     classes.append(folders)
  # print("classes", classes)
  # print("len of classes", len(classes))

  classes=[]
  for folders in os.listdir(dir):
    classes.append(folders)
  # print("classes", classes)
  # print("len of classes", len(classes))

  return classes

def encode_labels(classes_list):

  # label_encoder = preprocessing.LabelEncoder()
  # fit_t = label_encoder.fit_transform(classes)
  # print(fit_t)

  label_encoder = preprocessing.LabelEncoder()
  fit_t = label_encoder.fit_transform(classes_list)
  print(fit_t)

  return label_encoder

# train_image_paths, length_train_image_paths_list = generate_image_paths(train_dir)

test_image_paths, length_test_image_paths_list = generate_image_paths(test_dir)

classes = generate_classes_from_directory(train_dir)
label_encoder = encode_labels(classes)

class JerseyNumberDataset(Dataset):
    
    def __init__(self, image_paths, length, transform=None):
        self.image_paths = image_paths
        self.transform = transform
        self.length = length

    def __len__(self):
        # return 26010
        return self.length
    
    def __getitem__(self, index):
        #image_name = os.path.join(self.image_dir, self.image_files[index])
        #image = PIL.Image.open(image_name)
        #label = os.path.dirname(image_name)
        
        image_name = self.image_paths[index]
        #print("image name", image_name)

        image = PIL.Image.open(image_name)
        
        # image = cv2.imread(image_name)
        # print(type(image))
        # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        # print(type(image))
        # print(image)

        full_path_of_dir_containing_image = os.path.dirname(image_name)
        # print(full_path_of_dir_containing_image)
        label = os.path.basename(full_path_of_dir_containing_image)
        # print(type(label))
        label = label_encoder.transform([label])
        # print("label:", label)
        
        if self.transform:
            image = self.transform(image)
        return image,label
        
```

```python
# ==================== GETTING FROM GOOGLE COLAB ========================= 
# train_dir = "gdrive/My Drive/Colab Notebooks/jersey numbers dataset 224"
# ==================== GETTING FROM GOOGLE COLAB =========================

train_dir = "jersey numbers dataset 224"
test_dir = "test jersey numbers dataset w,h more than 100"

transform_train = transforms.Compose([
  # transforms.RandomHorizontalFlip(),
  transforms.Resize( (100, 100) ),
  # transforms.RandomCrop(100, padding=4),
  transforms.ToTensor(),
  transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
])

def generate_image_paths(dir):

  # train_image_paths = []
  # for folders in os.listdir(train_dir):
  #   for files in os.listdir(train_dir + '/' + folders):
  #     #print(files)
  #     train_image_paths.append(train_dir+'/'+folders+'/'+files)

  # print(train_image_paths[3])

  image_paths = []
  for folders in os.listdir(dir):
    for files in os.listdir(dir + '/' + folders):
      #print(files)
      image_paths.append(dir+'/'+folders+'/'+files)

  return image_paths, len(image_paths)

def generate_classes_from_directory(dir):
  # classes=[]
  # for folders in os.listdir(train_dir):
  #     classes.append(folders)
  # print("classes", classes)
  # print("len of classes", len(classes))

  classes=[]
  for folders in os.listdir(dir):
    classes.append(folders)
  # print("classes", classes)
  # print("len of classes", len(classes))

  return classes

def encode_labels(classes_list):

  # label_encoder = preprocessing.LabelEncoder()
  # fit_t = label_encoder.fit_transform(classes)
  # print(fit_t)

  label_encoder = preprocessing.LabelEncoder()
  fit_t = label_encoder.fit_transform(classes_list)
  print(fit_t)

  return label_encoder

# train_image_paths, length_train_image_paths_list = generate_image_paths(train_dir)

test_image_paths, length_test_image_paths_list = generate_image_paths(test_dir)

classes = generate_classes_from_directory(train_dir)
label_encoder = encode_labels(classes)

class JerseyNumberDataset(Dataset):
    
    def __init__(self, image_paths, length, transform=None):
        self.image_paths = image_paths
        self.transform = transform
        self.length = length

    def __len__(self):
        # return 26010
        return self.length
    
    def __getitem__(self, index):
        #image_name = os.path.join(self.image_dir, self.image_files[index])
        #image = PIL.Image.open(image_name)
        #label = os.path.dirname(image_name)
        
        image_name = self.image_paths[index]
        #print("image name", image_name)

        image = PIL.Image.open(image_name)
        
        # image = cv2.imread(image_name)
        # print(type(image))
        # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        # print(type(image))
        # print(image)

        full_path_of_dir_containing_image = os.path.dirname(image_name)
        # print(full_path_of_dir_containing_image)
        label = os.path.basename(full_path_of_dir_containing_image)
        # print(type(label))
        label = label_encoder.transform([label])
        # print("label:", label)
        
        if self.transform:
            image = self.transform(image)
        return image,label
        
```
